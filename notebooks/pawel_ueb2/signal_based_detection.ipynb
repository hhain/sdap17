# Exercise 3: Visualization

```python
# get necessary data

vals_2_4 = df_x1_t1_trx_1_4.loc[:,'trx_2_4_ifft_0':'trx_2_4_ifft_1999'].values
vals_1_4 = df_x1_t1_trx_1_4.loc[:,'trx_1_4_ifft_0':'trx_1_4_ifft_1999'].values
vals_3_4 = df_x1_t1_trx_1_4.loc[:,'trx_3_4_ifft_0':'trx_3_4_ifft_1999'].values

```

```python
# Spectral scheme
fig, axarr = plt.subplots(3, sharex=True)
axarr[0] = sns.heatmap(vals_2_4, xticklabels=200, yticklabels=20, vmin=0, vmax=1, cmap='Spectral', ax=axarr[0])
axarr[0].set_title('trx_2_4_ifft from measurement x1_t1_trx_1_4', fontsize=20)
axarr[1] = sns.heatmap(vals_1_4, xticklabels=200, yticklabels=20, vmin=0, vmax=1, cmap='Spectral', ax=axarr[1])
axarr[1].set_title('trx_1_4_ifft from measurement x1_t1_trx_1_4', fontsize=20)
axarr[2] = sns.heatmap(vals_3_4, xticklabels=200, yticklabels=20, vmin=0, vmax=1, cmap='Spectral', ax=axarr[2])
axarr[2].set_title('trx_3_4_ifft from measurement x1_t1_trx_1_4', fontsize=20)

plt.rcParams['figure.figsize'] = (20,20)
plt.show()
```
